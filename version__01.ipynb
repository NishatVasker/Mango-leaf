```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Sample dataset (replace this with your actual dataset)
X = np.random.rand(100, 5)  # 100 samples with 5 features each
y = np.random.randint(0, 2, size=100)  # Binary labels (0 or 1)

# Define the number of folds
k = 5

# Initialize k-fold cross-validation
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Lists to store the evaluation metrics for each fold
accuracy_scores = []

# Perform k-fold cross-validation
for fold, (train_index, test_index) in enumerate(kf.split(X)):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Initialize and train your model
    model = LogisticRegression()
    model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = model.predict(X_test)

    # Calculate accuracy for this fold
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

    print(f"Fold {fold+1}: Accuracy = {accuracy:.4f}")

# Calculate the average accuracy over all folds
average_accuracy = np.mean(accuracy_scores)
print(f"Average Accuracy: {average_accuracy:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for k-fold cross-validation
model_variants = ['YOLOv5s', 'YOLOv5m', 'YOLOv5l', 'YOLOv5x']
k_folds = 5

# Sample data for accuracy, precision, and recall for each fold and model variant
accuracy_data = np.array([[0.92, 0.88, 0.90, 0.86],
                          [0.90, 0.85, 0.87, 0.82],
                          [0.94, 0.90, 0.92, 0.88],
                          [0.91, 0.87, 0.89, 0.85],
                          [0.93, 0.89, 0.91, 0.87]])

precision_data = np.array([[0.90, 0.86, 0.88, 0.84],
                           [0.88, 0.83, 0.85, 0.80],
                           [0.92, 0.88, 0.90, 0.86],
                           [0.89, 0.85, 0.87, 0.83],
                           [0.91, 0.87, 0.89, 0.85]])

recall_data = np.array([[0.95, 0.92, 0.94, 0.90],
                        [0.92, 0.88, 0.90, 0.86],
                        [0.96, 0.93, 0.95, 0.91],
                        [0.94, 0.90, 0.92, 0.88],
                        [0.95, 0.92, 0.94, 0.90]])

# Create the ECG-like k-fold graph
plt.figure(figsize=(12, 6))
fold_values = np.linspace(0, 2 * np.pi, k_folds + 1)[:-1]
colors = ['b', 'g', 'r']
labels = ['Accuracy', 'Precision', 'Recall']

for i, metric_data in enumerate([accuracy_data, precision_data, recall_data]):
    for j, model in enumerate(model_variants):
        plt.plot(fold_values, metric_data[:, j], label=f'{model} - {labels[i]}', color=colors[i], linestyle='--')

plt.xlabel('Fold')
plt.ylabel('Metric Value')
plt.title('K-Fold Cross-Validation Performance')
plt.legend()
plt.xticks(np.linspace(0, 2 * np.pi, k_folds), range(1, k_folds + 1))
plt.grid(True)
plt.show()
```
